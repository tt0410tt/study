```python
import os
import random
import string
# 기본 파일생성 12분 53초
# 대형 파일 생성 함수 (랜덤 데이터 포함)
def create_large_file(file_path, size_gb):
    chunk_size = 1024 * 1024 * 10  # 10MB씩 쓰기
    total_size = size_gb * 1024 * 1024 * 1024  # GB를 바이트로 변환
    written = 0

    with open(file_path, "wb") as f:
        while written < total_size:
            # 10MB 크기의 랜덤 데이터를 생성하여 파일에 기록
            random_data = ''.join(random.choices(string.ascii_letters + string.digits, k=chunk_size)).encode('utf-8')
            f.write(random_data)
            written += chunk_size
            print(f"{written / (1024 * 1024):.2f} MB 생성 완료...", end="\r")

    print(f"\n파일 생성 완료: {file_path} ({size_gb}GB)")

# 소스 디렉토리 생성
source_dir = "source_files"
os.makedirs(source_dir, exist_ok=True)

# 대형 파일 생성 (각각 5GB)
file_count = 5
for i in range(1, file_count + 1):
    file_path = os.path.join(source_dir, f"large_file_{i}.dat")
    print(f"생성 중: {file_path}")
    create_large_file(file_path, 5)  # 각 파일 크기: 5GB
    print(f"완료: {file_path}")

print(f"총 {file_count}개의 대형 파일이 {source_dir} 폴더에 생성되었습니다.")
#12분 53초
```

```python
import os
import random
import string
import threading
# 13분 20초
# 대형 파일 생성 함수 (랜덤 데이터 포함)
def create_large_file(file_path, size_gb):
    chunk_size = 1024 * 1024 * 10  # 10MB씩 쓰기
    total_size = size_gb * 1024 * 1024 * 1024  # GB를 바이트로 변환
    written = 0

    with open(file_path, "wb") as f:
        while written < total_size:
            # 10MB 크기의 랜덤 데이터를 생성하여 파일에 기록
            random_data = ''.join(random.choices(string.ascii_letters + string.digits, k=chunk_size)).encode('utf-8')
            f.write(random_data)
            written += chunk_size
            print(f"{file_path}: {written / (1024 * 1024):.2f} MB 생성 완료...", end="\r")

    print(f"\n파일 생성 완료: {file_path} ({size_gb}GB)")

# 소스 디렉토리 생성
source_dir = "source_files"
os.makedirs(source_dir, exist_ok=True)

# 파일 생성 함수 (스레드용)
def create_file_task(file_index, size_gb):
    file_path = os.path.join(source_dir, f"large_file_{file_index}.dat")
    print(f"생성 중: {file_path}")
    create_large_file(file_path, size_gb)
    print(f"완료: {file_path}")

# 멀티스레딩으로 파일 생성
file_count = 5
file_size_gb = 5  # 각 파일 크기 (GB)
threads = []

# 스레드 생성 및 시작
for i in range(1, file_count + 1):
    thread = threading.Thread(target=create_file_task, args=(i, file_size_gb))
    threads.append(thread)
    thread.start()

# 모든 스레드 완료 대기
for thread in threads:
    thread.join()

print(f"총 {file_count}개의 대형 파일이 {source_dir} 폴더에 생성되었습니다.")
```

```python
import os
import random
import string
import multiprocessing
import time
#멀티코어 2분50초
# 대형 파일 생성 함수 (랜덤 데이터 포함)
def create_large_file(file_path, size_gb):
    chunk_size = 1024 * 1024 * 10  # 10MB씩 쓰기
    total_size = size_gb * 1024 * 1024 * 1024  # GB를 바이트로 변환
    written = 0

    with open(file_path, "wb") as f:
        while written < total_size:
            # 10MB 크기의 랜덤 데이터를 생성하여 파일에 기록
            random_data = ''.join(random.choices(string.ascii_letters + string.digits, k=chunk_size)).encode('utf-8')
            f.write(random_data)
            written += chunk_size
            print(f"{file_path}: {written / (1024 * 1024):.2f} MB 생성 완료...", end="\r")

    print(f"\n파일 생성 완료: {file_path} ({size_gb}GB)")

# 소스 디렉토리 생성
source_dir = "source_files"
os.makedirs(source_dir, exist_ok=True)

# 멀티프로세싱 작업 함수
def create_file_task(file_index, size_gb):
    file_path = os.path.join(source_dir, f"large_file_{file_index}.dat")
    print(f"생성 중: {file_path}")
    create_large_file(file_path, size_gb)
    print(f"완료: {file_path}")

if __name__ == "__main__":
    start_time = time.time()

    # 파일 생성 작업 정의
    file_count = 5
    file_size_gb = 5  # 각 파일 크기 (GB)
    processes = []

    # 프로세스 생성 및 시작
    for i in range(1, file_count + 1):
        process = multiprocessing.Process(target=create_file_task, args=(i, file_size_gb))
        processes.append(process)
        process.start()

    # 모든 프로세스 완료 대기
    for process in processes:
        process.join()

    end_time = time.time()
    print(f"총 {file_count}개의 대형 파일이 {source_dir} 폴더에 생성되었습니다.")
    print(f"멀티프로세싱 실행 시간: {end_time - start_time:.2f}초")
```
