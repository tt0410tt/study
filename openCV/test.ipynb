```python
point1 = np.float32([
    [150, 250],  # 왼쪽 위
    [450, 250],  # 오른쪽 위
    [115, 380],  # 왼쪽 아래
    [480, 380]   # 오른쪽 아래
])

# 변환 후 이미지에서의 좌표 (직사각형)
point2 = np.float32([
    [0, 0],
    [600, 0],
    [0, 300],
    [600, 300]
])

print(img.shape)
matrix=cv2.getPerspectiveTransform(point1,point2)
warp_image= cv2.warpPerspective(img,matrix,(800,500))
plt.figure()
plt.imshow(warp_image)
plt.title("warp")
plt.show()
plt.close()
```

```python
warped= cv2.warpPerspective(img,matrix,(900,320))
gray=cv2.cvtColor(warped,cv2.COLOR_BGR2GRAY)
blurred=cv2.GaussianBlur(gray,(5,5),0)
edges=cv2.Canny(blurred,50,150)
lines=cv2.HoughLinesP(edges,1,np.pi/180,threshold=100,minLineLength=50,maxLineGap=50)
line_image=np.copy(warped)

if lines is not None:
    for line in lines:
        x1,y1,x2,y2=line[0]
        cv2.line(line_image,(x1,y1),(x2,y2),(0,255,0),2)#녹색선

plt.figure()
plt.subplot(1,3,1)
plt.imshow(img)
plt.title("원본")

plt.subplot(1,3,2)
plt.imshow(edges,cmap='gray')
plt.title("캐니엣지")

plt.subplot(1,3,3)
plt.imshow(line_image, cv2.COLOR_BGR2RGB)
plt.title("모름")

plt.show()
plt.close()
```
