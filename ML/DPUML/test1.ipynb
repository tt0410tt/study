```python
import zipfile
import pandas as pd
# 압축 해제
with zipfile.ZipFile('disease-prediction-using-machine-learning.zip', 'r') as zip_ref:
    zip_ref.extractall('data_folder')

# 데이터 불러오기
train_data = pd.read_csv('data_folder/Training.csv')
test_data = pd.read_csv('data_folder/Testing.csv')

# 데이터 확인
print(train_data.head())
print(test_data.head())
```

```python
missing_cols_in_test = set(train_data.columns) - set(test_data.columns)
missing_cols_in_train = set(test_data.columns) - set(train_data.columns)

if missing_cols_in_test:
    print("\nTesting 데이터에 없는 열:", missing_cols_in_test)
else:
    print("\nTesting 데이터에 모든 Training 열이 존재합니다.")

if missing_cols_in_train:
    print("\nTraining 데이터에 없는 열:", missing_cols_in_train)
else:
    print("\nTraining 데이터에 모든 Testing 열이 존재합니다.")

# 5. null 값 확인
print("\nTraining 데이터의 Null 값 확인:")
print(train_data.isnull().sum())

print("\nTesting 데이터의 Null 값 확인:")
print(test_data.isnull().sum())
```

```python
# Unnamed 열 이름 확인
if 'Unnamed: 133' in train_data.columns:
    print("\nTraining 데이터의 133번째 열(Unnamed) 내용:")
    print(train_data['Unnamed: 133'].head())
else:
    print("\nTraining 데이터에 'Unnamed: 133' 열이 없습니다.")

if 'Unnamed: 133' in test_data.columns:
    print("\nTesting 데이터의 133번째 열(Unnamed) 내용:")
    print(test_data['Unnamed: 133'].head())
else:
    print("\nTesting 데이터에 'Unnamed: 133' 열이 없습니다.")
```

```python
# 특정 열의 NaN 개수 확인
if 'Unnamed: 133' in train_data.columns:
    nan_count = train_data['Unnamed: 133'].isnull().sum()
    print(f"'Unnamed: 133' 열의 NaN 개수: {nan_count}")
    print(f"'Unnamed: 133' 열의 전체 행 개수: {len(train_data)}")
```

```python
x_train = train_data.drop(columns=['prognosis','Unnamed: 133'])
y_train = train_data[['prognosis']]
x_test = test_data.drop(columns='prognosis')
y_test = test_data[['prognosis']]
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear')
print(y_train)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

혼재 시켜보자 1.0이 안나오도록

```python
from sklearn.model_selection import train_test_split
import numpy as np
# 데이터 준비
Train_Data = train_data.drop(columns=['prognosis', 'Unnamed: 133'])
Train_Labels = train_data[['prognosis']]
Test_Data = test_data.drop(columns='prognosis')
Test_Labels = test_data[['prognosis']]

# 데이터 합치기
X = pd.concat([Train_Data, Test_Data], axis=0)
Y = pd.concat([Train_Labels, Test_Labels], axis=0)

# 합친 데이터 다시 나누기
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model2 = SVC(kernel='linear')
model2.fit(X_Train,Y_Train)

Y_pred = model2.predict(X_Test)
print("Accuracy:", accuracy_score(Y_Test, Y_pred))
```
